### 14. What trade-offs are consumers willing to make between price, health benefits, and taste?

```python
df_data[["Price_range","Health_concerns","Taste_experience"]]
```

```python
data = df_data[["Price_range","Health_concerns","Taste_experience"]]
```

```python
import pandas as pd
import scipy.stats as stats

# Create contingency tables
price_health = pd.crosstab(data['Price_range'], data['Health_concerns'])
price_taste = pd.crosstab(data['Price_range'], data['Taste_experience'])

# Chi-square test between price and health benefits
chi2, p, _, _ = stats.chi2_contingency(price_health)
print(f"Price vs Health Benefits p-value: {p}")

# Chi-square test between price and taste
chi2, p, _, _ = stats.chi2_contingency(price_taste)
print(f"Price vs Taste p-value: {p}")

# Interpret significance
if p < 0.05:
    print("There is a significant trade-off between the variables.")
else:
    print("No significant trade-off found.")
```
